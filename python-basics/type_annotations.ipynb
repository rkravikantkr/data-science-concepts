# Type Annotations
- Python is dynamically typed language.
- type annotations doesn't force static typing
- type annotations are helpful in making program informative and more explicit
- 

```python
# type of a and b are ambiguous, so is the returned value
def add(a,b):
    return a+b
    
#  "+" is valid for integers, floats, strings, list.
assert add(1,1) == 2 
assert add([1],[2,3]) == [1,2,3]
assert add("a","b") =="ab"
assert add(1.3,3.7) == 5.0
assert add(1.5, 5) == 6.5

try:
    add(1,"1")
except TypeError:
    print("arguments of different types cannot be added")
```

```python
# more explicit defination of add function
# 
def add(a:int, b:int) -> int: # return type is of same type as parameter
    return a+b

assert add(1,3) == 4
assert add("a","b") == "ab" # still add fn can add two strings.
# having type annotations makes the function more explicity and readable

# def dot_product(x,y):...
# def dot_product(x:Vector, y:Vector):...
```

# typing module provides parameterized types
- from typing import List
- variables can also be type annotated
- funtions in python are first-class objects, they can also be type-annotated

```python
from typing import List, Union

def average(l:List[int]) -> float:
    result = 0
    for x in l:
        result += x
    return result/len(l)

assert average([1,1]) == 1
```

```python
from typing import Callable

# repeater is a fn with two parametes of str and int type respectively and returns a str

def twice_repeat(repeater: Callable[[str, int], str], s: str) -> str:
    return repeater(s, 2)

def repeat_str(s: str, n: int) -> str:
    n_copies = [s for _ in range(n)]
    return ''.join(n_copies)

print(twice_repeat(repeat_str, "_"))  # Output: a, a
print(repeat_str("*",100)) # 
```
