# Vectors
- Physics: A vector is a quantity that has both a magnitude(size) and a direction.
- How much and which way.
- Maths: A vector is a ordered tuple of real numbers. v = (v1, v2, v3 ..., vn), each vi is a component along a coordinate axis, in  n-dimensional real space.

## The Formal Definition (Algebraic)
- A vector space V over a field F is a set with two operations
- Addition: u + v ∈ V
- Scalar multiplication: c·v ∈ V
- That satisfy 8 axioms for all u, v, w ∈ V and scalars a, b ∈ F:
- 1) Commutativityu + v = v + u (2)Associativity(u + v) + w = u + (v + w), (3)Zero vectorv + 0 = v (4)Additive inversev + (−v) = 0 (5)Scalar identity1·v = v (6)Scalar associativitya(bv) = (ab)v (7)Distributivity 1a(u + v) = au + av (8)Distributivity 2(a + b)v = av + bv

### Any object satisfying these axioms is a vector — regardless of what it looks like.
```
Magnitude (length) of a vector v = (v₁, v₂, ..., vₙ):
‖v‖ = √(v₁² + v₂² + ... + vₙ²)
Dot product of u and v:
u · v = u₁v₁ + u₂v₂ + ... + uₙvₙ = ‖u‖ ‖v‖ cos θ
where θ is the angle between them.
```

````python
# Addition of two vectors
# vectors add componentwise
# Two vectors of different lengths belong to fundamentally different spaces.
# same dimension -> can be added

from typing import List
Vector = List[float]

def add(v:Vector, w:Vector) -> Vector:
    """Adds corresoponding elements"""
    assert len(v) == len(w) # vectors must be the length
    return [vi + wi for vi, wi in zip(v,w)] # vi / wi are i_th component

a = [1,2,3,9]
b = [4,5,6,0]
c = [5,7,9,9]
assert add(a,b) == c 

def subtract(v:Vector, w:Vector) -> Vector:
    """Subtract corresponding elements"""
    assert len(v) == len(w)
    return [vi - wi for vi, wi in zip(v,w)]

assert subtract(a,b) == [-3,-3,-3,9]
````

````python
# Componentwise sum of a list of vectors
def vector_sum(vectors:List[Vector]) -> Vector:
    """Sums all corresponding elements"""
    # check if vectors is not empty
    assert vectors, "no vectors provided!"

    # check if all vectors are of same length
    length = len(vectors[0])
    assert all(len(v) == length for v in vectors), "vectors have different sizes!"

    # i_th element is sum of all vector[i]
    return [sum(vector[i] for vector in vectors)
            for i in range(length)]

assert vector_sum([[1,2],[1,2]]) == [2,4]
````

````python
# Scalar multiplication of a vector
def scalar_multiply(scalar:float, vector:Vector) -> Vector:
    """Multiplies scalar/constant scalar with every element """
    return [scalar*vi for vi in vector]

assert scalar_multiply(scalar=2,vector=[1,2]) == [2,4]
````

````python
# Dot product of two same sized vectors
def dot(v:Vector, w:Vector)->float:
    assert len(v)==len(w)
    return sum(vi*wi for vi, wi in zip(v,w))

assert dot([1,2,3],[4,5,6]) == 32
````

````python
# vector's sum of squares
def sum_of_squares(v:Vector)->float:
    """returns v1*v1 + v2*v2....vn*vn"""
    return dot(v,v)

assert sum_of_squares([1,2,3]) == 14
````

````python
# magnitude / length of a vector
import math
def magnitude(v:Vector)->float:
    return math.sqrt(sum_of_squares(v))

assert magnitude([3,4])==5
````

````python
# distance between two vectors

def squared_distance(v:Vector, w:Vector)->float:
    return sum_of_squares(subtract(v,w))

def distance(v:Vector, w:Vector) -> float:
    # return math.sqrt(squared_distance(v,w))
    return magnitude(subtract(v, w))
````
